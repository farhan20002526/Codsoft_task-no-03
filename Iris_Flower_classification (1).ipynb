```python
dummies=pd.get_dummies(df,columns=["species"])
dummies.head()

```

```python
df_m=dummies
```

```python
df_m.head()
```

```python
columns_to_keep = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
X = df[columns_to_keep]
```

```python
Y=df_m.drop(columns=columns_to_keep)
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
len(X_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

```python
model.fit(X_train,Y_train)
```

```python
model.predict(X_test)
```

```python
import numpy as np

def predict_flr(sepal_length, sepal_width, petal_length, petal_width):
    t = np.zeros(len(X.columns))


    t[0] = sepal_length
    t[1] = sepal_width
    t[2] = petal_length
    t[3] = petal_width


    predicted_values = model.predict([t])[0]


    flower_names = ['setosa', 'versicolor', 'virginica']
    predicted_flower = flower_names[np.argmax(predicted_values)]

    return predicted_flower
```

```python
predicted_flower = predict_flr(5.1, 3.5, 1.4, 0.2)
print(predicted_flower)
```

```python
model.score(X,Y)
```
